```python
dict1={'Customer_id':[1,2,3,4,5,6],'Product':['Television','Television','Television','Earphones','Earphones','Earphones']}
df1=pd.DataFrame(dict1)
df1
```

```python
dict2={'Customer_id':[2,4,6],'State':['California','California','Texas']}
df2=pd.DataFrame(dict2)
df2
```

```python
inner_join_df = pd.merge(df1, df2, on='Customer_id', how='inner')
inner_join_df
```

```python
outer_join_df = pd.merge(df1, df2, on='Customer_id', how='outer')
outer_join_df
```
